```python
import pandas as pd
from pycovid import pycovid
import numpy as np

import seaborn as sns
```

```python
df = pycovid.getCovidCases(countries=["Austria", "Spain", "Germany", "Italy", "Switzerland", "Iran"])
df['month'] = df['date'].map(lambda x: x.month)
```

```python
x = df[(df.type=="death")].groupby(['country_region', 'month'])['cases'].sum()
x.unstack()
```

```python
mort = np.arange(.01, .061, .01)
inf = np.arange(.6, .81, .1)

df = pd.DataFrame()
for m in mort:
    for i in inf:
        d = 9000000*m*i
        df.loc[int(m*100), i*100] = d

df.index.name = "Mortalität (%)"
df.columns = [f"{int(_)}%" for _ in df.columns]
```

```python
cm = sns.light_palette("green", as_cmap=True)
df.astype(int).style.background_gradient(cmap=cm, axis=None)
```
